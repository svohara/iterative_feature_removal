## Using IFR on your own data

To make using the ifr package as easy as possible, we recommend that you
add it to your PYTHON_PATH (in windows, the PythonPath environment
variable), or install to your site_packages directory. Directions on how
to do so can be found on the web.

Assuming the ifr package has been correctly added to the appropriate
path, you can import it into your own projects, or use interactively
with the interpreter. I highly recommend using the ipython interpreter
over the base python interpreter.

Here is an example of computing IFR on a given data set. We assume you
have the ability to load your data into a python matrix. Look at the
scipy (or numpy) function loadtxt(...) if you want to load data stored
in text/csv files.

```python
import ifr
```

```python
# load your data matrix, D, here D should have rows as samples, features
# as columns load your labels as well as a vector of integers, L note:
# D.shape[0] == len(L)
# 
# Also (optional) load feature names as list variable F, len(F) ==
# D.shape[1], so each column has a string name in F
# 
# for the sake of demonstration, we'll load the H3N2 influenza data
(D,L,T,S) = ifr.load_flu_mat()
# NOTE: S is not used here, but it is the subject identifier for each
# sample NOTE: T is not used either, and is the time interval for each
# sample
# 
#gene names corresponding to Duke data
F = ifr.load_gene_ids(short_name_only=True)  
```

```python
F[0:10]
```

```python
D.shape
```

```python
len(L)
```

```python
#note: we can partition your data for you using functions
# found in the cross_validation module included in the ifr package
(D_train,L_train,D_test,L_test,_) = ifr.random_partition(D, L)
traindat = (D_train, L_train)  #grouping for convenience
testdat = (D_test, L_test)
```

```python
#now create an IFR object using your data the following uses defaults,
#but you may need to pass additional keyword parameters to tune the
#machine learning to your data. NOTE: do NOT name the ifr object 'ifr'
#because that is the package name and things will break. Choose another
#variable name, like I
I = ifr.IFR( traindat, testdat, feature_names=F)
```

```python
I.computeAllIterations()
```

```python
I.plotResults()
```

```python
I.export_iterations(fn="iterations.txt")
```

## Using Pathway Classification on your own data

If your data has feature names which are genes, and can be queried using
the GATHER web interface (gather.genome.duke.edu), then you may wish to
see if organizing the features from the first N iterations of IFR can
generate discriminative classifiers. Choose N by looking at the results
of the plotResults() method and determining at which point does the
rolling average of the test accuracy fall below some acceptable value.

```python
#Continuing from above example, where object I is an IFR object with the
#iterations already computed.

#Create a new pathway classifier object
pc = ifr.Pathway_Classifier( traindat, testdat, feature_names=F,
                                numIter=30, ifr_obj=I)
```

```python
#Since we already have a pre-computed IFR object, we just need to tell
#the pathway classifier to perform pathway analysis using a built-in IFRA
# object (IFR Analysis object), initialized with the appropriate values.
# We do this by passing in a dictionary of parameter settings to the
# IFRA_kwargs argument of the initAnalysis method. In this case, we use
# all defaults, and set the GO ontololgy depth to 5.
#NOTE: This requires an internet connection, and it is not the fastest
#thing in the world...
pc.initAnalysis( IFRA_kwargs={'depth':5} )
```

NOTE: in the above output, the occassional error saying "there are no annotations"
    is normal. Due to how the gene ontology is organized, not all genes will have
    annotations at all levels, and thus it is possible to have an iteration with
    no pathway annotations.

```python
#Once the annotations have been fetched, we can compute the top pathway
# and pathway pair classifiers as follows.
#min_size=5 will ignore pathways that have less than 5 genes represented
rc = pc.computePathwayClassifiers(min_size=5)
```

```python
#To get a list of the top pathway classifiers do the following
pc.getTopPathways()  # will return the top-10, but you can change
```

```python
#Similarly for the top pairs
rc = pc.computePathwayPairsClassifiers(verbose=False)
pc.getTopPathwayPairs(18)  #return the top 18 in this case...
```

```python
#Find the genes from a given pathway in the pathway classifier
pw = pc.getTopPathways(1)[0][1]
print "Best pathway is: %s"%pw  #well...it may be tied for best
gl = pc.getGenelistForPathway(pw)
print "Which consists of the following genes:"
print gl
```

```python
#Load some independent test data, using the H1N1 data
(Dx,Lx,_,_) = ifr.load_H1N1_mat()
```

```python
#Does the genelist from the 'best' pathway have features that will
# separate the H1N1 data?

#NOTE: The following function is a convenience wrapper of the ifr.svm_engine,
# and it defaults to parameters which seem to work well on the data sets we
# explored in our paper. It automatically mean-centers the data and scales
# each feature to unit deviation. The setting common_scale=False means that
# the test data is scaled independently of the training. This setting should
# be True if working with the Lung Cancer data because the test data is highly
# uneven in distribution, but the trainind data is even between the classes.
rc = ifr.pathway_classification(gl, feature_names=F, traindat=(D,L), testdat=(Dx,Lx), common_scale=False)
print "Classification accuracy on the test data is %3.2f percent"%(100*rc[0])
print "Using genes from: %s"%pw
```

```python
#How do we get the gene list of the best pathway pair?
rc = pc.getTopPathwayPairs(1)
pair_str = rc[0][1]
(p1,p2) = pair_str.split(" + ")
print "Pathway 1 is %s"%p1
print "Pathway 2 is %s"%p2
```

```python
gl1 = pc.getGenelistForPathway(p1)
gl2 = pc.getGenelistForPathway(p2)
gl3 = gl1 + gl2
print gl3
```

```python
#it should classify the test data at least as well as taxis alone...
rc = ifr.pathway_classification(gl3, feature_names=F, traindat=(D,L), testdat=(Dx,Lx), common_scale=False)
print "Classification accuracy on the test data is %3.2f percent"%(100*rc[0])
print "Using genes from: %s"%pair_str
```

#### NOTE: If you re-run all the tutorial code, you will be generating a new random split of the training data, so your final results on the pathway classification will vary. Although if you run many trials, you will find certain pathways are likely to be identified as highly discriminative, and thus it is possible to choose pathways based on multiple-trial cross-validation, as we've done for our paper. Also, remember that this tutorial does not split the data to prevent subject overlap so the training/validation results are overly optimistic. We have a more-involved procedure for multiple-trial cross validation, which you can find in the scripts/bmc_genomics_paper.py script module.
